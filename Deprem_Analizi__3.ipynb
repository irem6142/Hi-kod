```python
for i,datetime in enumerate(df["eventDate"],0):
    datetime_split = datetime.split("T")
    date_split = datetime_split[0].split("-")
    df["eventDate"].iloc[i] = date_split[0]+"/"+date_split[1]+"/"+date_split[2] + " " + datetime_split[1]
```

```python
for i,date in enumerate(df["eventDate"],0):
    df["eventDate"].iloc[i] = pd.to_datetime(date)
    
df.head()
```

#  Kahramanmaraş Deprem Analizi

```python
maras_index=[]
for i,sehir in enumerate(df["area"],0):
    if "Kahramanmaraş" in df["area"].loc[i]:
        maras_index.append(i)
```

```python
data_maras = df.iloc[maras_index]
data_maras
```

```python
for i,date in enumerate(data_maras["eventDate"],0):
    data_maras["eventDate"].iloc[i] = pd.to_datetime(date)
    
df.head()
```

```python
plt.figure(figsize = (15,5))
plt.style.use("seaborn-whitegrid")
plt.title("KAHRAMANMARAŞ DEPREM BÜYÜKLÜĞÜ - TARİH ANALİZİ")
plt.xlabel("DEPREM TARİHLERİ")
plt.ylabel("DEPREM BÜYÜKLÜKLERİ")
sns.lineplot(x = "eventDate", y = "magnitude", data =  data_maras)
```

```python
plt.figure(figsize = (15,5))
plt.style.use("seaborn-whitegrid")
plt.title("KAHRAMANMARAŞ DEPREM DERİNLİĞİ - TARİH ANALİZİ")
plt.xlabel("DEPREM TARİHLERİ")
plt.ylabel("DEPREM DERİNLİKLERİ")
sns.lineplot(x = "eventDate", y = "depth", data =  data_maras)
```
